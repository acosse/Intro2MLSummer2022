### Additional note on Gaussian Discriminant Analysis

#### I) The multivariate Normal Distribution

```python
# Simple illustration of the MVN
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm


x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)

xx1, xx2 = np.meshgrid(x1,x2)

X = np.vstack((xx1.flatten(), xx2.flatten())).T


# multivariate Gaussian


mean = np.asarray([1,1])

Sigma = np.eye(2,2)
D = 2


MVN = np.zeros(np.shape(X)[0],)

for i in np.arange(np.shape(X)[0]):
    current = X[i,:]
    tmpL = np.matmul(current-mean,np.linalg.inv(Sigma))
    tmp = np.matmul(tmpL, (current-mean).T)
    MVN[i] = 1/((2*np.pi)**(D/2) * np.linalg.det(Sigma)*(1/2))\
    * np.exp(-(1/2)*(tmp)) 
    
    
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(xx1, xx2, np.reshape(MVN, np.shape(xx1)), \
                alpha=0.5, cmap=cm.get_cmap('hsv'))
plt.show()
```

#### II) LDA as a logistic regression classifier

```python
# Connection between LDA and Logistic regression

# Generating binary dataset 

import numpy as np
import matplotlib.pyplot as plt

mu1, sigma1 = -1, 0.5 
x_class1 = np.random.normal(mu1, sigma1, 30)

mu2, sigma2 = 1, 0.5 
x_class2 = np.random.normal(mu2, sigma2, 30)

plt.scatter(x_class1, np.zeros(np.shape(x_class1)), c='b')
plt.scatter(x_class2, np.zeros(np.shape(x_class2)), c='r')
plt.show()
```

```python
pi_1 = len(x_class1)/(len(x_class1) + len(x_class2))

mu1_recovered = np.mean(x_class1)
mu2_recovered = np.mean(x_class2)

tmp1 = x_class1 - mu1_recovered
tmp2 = x_class2 - mu2_recovered


covariance = (1/(len(x_class1) + len(x_class2)))*\
(np.sum(tmp1**2)+ np.sum(tmp2**2))


x = np.linspace(-3,3,100)


gaussian_class1 = (1/np.sqrt(2*np.pi)/np.sqrt(covariance))*\
np.exp(-(1/2)*(x - mu1_recovered)**2/covariance)

gaussian_class2 = (1/np.sqrt(2*np.pi)/np.sqrt(covariance))*\
np.exp(-(1/2)*(x - mu2_recovered)**2/covariance)


plt.scatter(x_class1, np.zeros(np.shape(x_class1)), c='b')
plt.scatter(x_class2, np.zeros(np.shape(x_class2)), c='r')
plt.plot(x, gaussian_class1, c='b')
plt.plot(x, gaussian_class2, c='r')
plt.show()

```

```python
# p(t = 1|x) = (p(x|t = 1)p(t=1)) / (p(x|t=1)p(t=1) + p(x|t=2)p(t=2))

posterior = np.divide(gaussian_class1*pi_1, \
                      gaussian_class1*pi_1 + gaussian_class2*(1-pi_1)) 

plt.scatter(x_class1, np.zeros(np.shape(x_class1)), c='b')
plt.scatter(x_class2, np.zeros(np.shape(x_class2)), c='r')
plt.plot(x, gaussian_class1, c='b')
plt.plot(x, gaussian_class2, c='r')
plt.plot(x, 1-posterior, c='g')
plt.show()
```
